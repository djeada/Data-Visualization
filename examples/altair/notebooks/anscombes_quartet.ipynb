```python
import altair as alt
import numpy as np
import pandas as pd

# Define the dataset
dataset_1 = np.array([
    [10, 8.04], [8, 6.95], [13, 7.58], [9, 8.81], [11, 8.33],
    [14, 9.96], [6, 7.24], [4, 4.26], [12, 10.84], [7, 4.82], [5, 5.68]
])

# Create DataFrames for each dataset
df_1 = pd.DataFrame(dataset_1, columns=['x', 'y'])

# Scatter plots for each dataset
alt.Chart(df_1).mark_circle().encode(
    x='x',
    y='y'
)
```
