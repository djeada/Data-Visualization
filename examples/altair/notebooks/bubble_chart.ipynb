```python
# Import libraries
import altair as alt
import numpy as np
import pandas as pd

# Generate the data
dfs = []

for i in range(3):
    x = np.linspace(0, 10, 100)
    y = 2 * x + 1 + np.random.randn(100) * 2
    z = np.random.normal(0.5, 2, 100)
    df = pd.DataFrame({"x": x, "y": y, "size": z * 100, "dataset": f"data {i + 1}"})
    dfs.append(df)

# Combine all dataframes
df_all = pd.concat(dfs)

# Create a bubble chart
alt.Chart(df_all).mark_circle().encode(
    x="x",
    y="y",
    size=alt.Size(
        "size", scale=alt.Scale(range=[0, 1000]), legend=alt.Legend(title="Size")
    ),
    color="dataset:N",
    tooltip=["x", "y", "size", "dataset"],
).properties(width=700, height=400)
```
