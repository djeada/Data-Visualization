```python
import altair as alt
import pandas as pd
import numpy as np

# Create some data
np.random.seed(0)
x = np.random.normal(0, 1, 3000)
y = np.random.normal(0, 1, 3000)
df = pd.DataFrame({"x": x, "y": y})

# Overplotted scatter plot
alt.Chart(df).mark_circle(opacity=0.3).encode(x="x", y="y").properties(
    width=600, height=300
)
```

```python
# 2D density plot
alt.Chart(df).mark_rect().encode(
    alt.X("x:Q", bin=alt.Bin(maxbins=30)),
    alt.Y("y:Q", bin=alt.Bin(maxbins=30)),
    alt.Color("count()", scale=alt.Scale(scheme="blues")),
).properties(width=600, height=300)
```
