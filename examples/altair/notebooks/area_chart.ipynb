```python
import altair as alt
import pandas as pd
import numpy as np

x = np.arange(1, 6)
y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([2, 3, 4, 5, 6])
y3 = np.array([3, 4, 5, 6, 7])

# Create a DataFrame
df = pd.DataFrame({"x": x, "y1": y1, "y2": y2, "y3": y3})

# Transform the DataFrame into long format
df = df.melt("x", var_name="series", value_name="value")

# Create the area chart
chart = (
    alt.Chart(df)
    .mark_area(opacity=0.1)
    .encode(x="x", y="value", color="series")
    .properties(title="Area Chart")
)

# Set the axis labels
chart.configure_axis(labelFontSize=12, titleFontSize=14).encode(
    alt.X("x", title="Time"), alt.Y("value", title="Value")
)
```
