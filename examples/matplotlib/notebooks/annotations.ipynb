```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)

# create sine plot
x = np.arange(0, 10, 0.1)
y = np.sin(x)
ax.plot(x, y)

# add annotation with arrow for highest and lowest values
index_max = np.argmax(y)
index_min = np.argmin(y)
ax.annotate(
    "highest",
    xy=(x[index_max], y[index_max]),
    xytext=(x[index_max] - 0.5, y[index_max] - 0.5),
    arrowprops=dict(facecolor="black", shrink=0.05),
)
ax.annotate(
    "lowest",
    xy=(x[index_min], y[index_min]),
    xytext=(x[index_min] - 0.5, y[index_min] + 0.5),
    arrowprops=dict(facecolor="black", shrink=0.05),
)

plt.show()
```
