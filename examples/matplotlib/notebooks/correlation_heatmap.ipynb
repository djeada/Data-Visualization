```python
import pandas as pd
import numpy as np
from typing import Optional
import matplotlib.pyplot as plt
```

```python
def plot_correlation_heatmap(data_frame: pd.DataFrame, title: Optional[str]=None) -> None:
    """
    Plots a correlation heatmap for a given list of columns.

    :param data_frame: the data frame holding the data.
    :param title: title of the plot.
    """
    corr = data_frame.corr()
    fig = plt.figure(figsize=(8, 8))
    ax = plt.subplot(111,aspect='equal')    
    plt.subplots_adjust(left=0.05, right=1.1, top=0.85, bottom=0.05)
    ax = ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), labels = corr.columns, rotation=45)
    plt.yticks(range(len(corr.columns)), labels = corr.columns)
    fig.colorbar(ax)
    if title is not None:
        fig.set_title(title)
    plt.show()
```

```python
# create a higly correlated data set
n = 100
df = pd.DataFrame(np.random.randn(n, 2), columns=['A', 'B'])
df['C'] = df['A'] + np.random.randn(n) 
df['D'] =  df['B'] + np.random.uniform(low=0.5, high=33.3, size=(n,))
df['E'] = np.random.poisson(5, size=(n,))
df['F'] = np.sin(n) + df['B']

plot_correlation_heatmap(df)
```
