```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
df = pd.DataFrame(
    {"A": [1, 2, 3, 4, 5], "B": [2, 3, 4, 5, 6], "C": [3, 4, 5, 6, 7]},
    index=[1, 2, 3, 4, 5],
)

df.plot(kind="area", stacked=True)

plt.title("Stacked Area Chart")
plt.xlabel("Time")
plt.ylabel("Value")

plt.show()
```
