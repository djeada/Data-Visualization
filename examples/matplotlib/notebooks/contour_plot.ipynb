```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)

x_list = np.linspace(-5, 5, 100)
y_list = np.linspace(-5, 5, 100)

x, y = np.meshgrid(x_list, y_list)
z = np.sqrt(x**2 + y**2)
    
contour_plot = ax.contourf(x, y, z)
fig.colorbar(contour_plot)
plt.show()
```
