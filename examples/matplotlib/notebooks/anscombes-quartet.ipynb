```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset_1 = np.array(
    [
        [10, 8.04],
        [8, 6.95],
        [13, 7.58],
        [9, 8.81],
        [11, 8.33],
        [14, 9.96],
        [6, 7.24],
        [4, 4.26],
        [12, 10.84],
        [7, 4.82],
        [5, 5.68],
    ]
)

x = dataset_1[:, 0]
y = dataset_1[:, 1]

plt.scatter(x, y)

regression_x = np.linspace(min(x), max(x), 100)
regression_y = 0.5 * regression_x + 3

plt.plot(regression_x, regression_y, "orange")
plt.show()
```
