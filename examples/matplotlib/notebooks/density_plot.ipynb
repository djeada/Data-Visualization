```python
import matplotlib.pyplot as plt
import numpy as np

data = np.zeros((100,100))
for i in range(100):
    for j in range(100):
           data[i,j] = (i-50) + (j-50)
                   
fig = plt.figure()
ax = fig.add_subplot(111)
density_plot = ax.imshow(data, interpolation='nearest')
plt.colorbar(density_plot)
plt.show()
```
