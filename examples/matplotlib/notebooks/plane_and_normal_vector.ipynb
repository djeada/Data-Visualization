```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xy = np.mgrid[-2:2:21j, -2:2:21j]
z = np.zeros((21, 21))
x = xy[0]; y = xy[1]

d = 1.0
z = d - x - y
ax.plot_surface(x, y, z, cmap=cm.coolwarm, linewidth=0, alpha=.7)

# Normal vector
a = [0, 1]; b = [0, 1]; c = [0, 1]
ax.plot(a, b, c, color="blue")

ax.set_title('A Plane in 3D Defined by a Normal Vector')
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
ax.set_xlim(-2, 2); ax.set_ylim(-2, 2); ax.set_zlim(-2, 2)

# Try plotting with and without the next line to see what it can do for us.
ax.set_box_aspect((np.ptp(x), np.ptp(y), np.ptp(x)))
ax.tick_params(axis='both', which='major', labelsize=6)

plt.show()
```
