```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate the data

datasets = []

for i in range(3):
    x = np.linspace(0, 10, 100)
    y = 2 * x + 1 + np.random.randn(100) * 2
    z = np.random.normal(0.5, 2, 100)
    datasets.append([x, y, z])
```

```python
# plot the data
plt.figure(figsize=(10, 5))
for x, y, z in datasets:
    plt.scatter(x, y, s=z * 100, alpha=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Bubble plot")
plt.legend(["data 1", "data 2", "data 3"])
plt.show()
```
